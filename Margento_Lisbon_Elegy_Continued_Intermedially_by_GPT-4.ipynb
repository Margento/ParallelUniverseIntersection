```python

def generate_videopoem_prompt(poem_sequence, motif_path, emotional_progression):
    """
    Compose a GPT-4 prompt that instructs the model to generate a coherent videopoem based on motifs, coherence principles,
    and existing poem context.

    Args:
        poem_sequence (str): All previous poems in the Lisbon. An Inter-M√∂bius Elegy sequence.
        motif_path (list[str]): Ordered list of motifs extracted from graph analysis.
        emotional_progression (str): The desired narrative arc.

    Returns:
        str: Complete prompt string ready to send to GPT-4.
    """

    coherence_explanation = """
You are a videopoem creator who blends poetry, voice, and image into unified lyrical works.

Your goal is to continue a poetic sequence titled *Lisbon. An Inter-M√∂bius Elegy* by producing a new intermedia piece with high audio-video coherence, as defined by:

- **Micro-transition coherence**: Sudden scene or rhythm shifts should be mirrored across sound and image.
- **Mid-range fluidity**: Each group of lines/scenes should remain stylistically and emotionally consistent.
- **Local harmony**: Match speech pacing, pauses, and tempo with scene rhythm and visual intensity.
- **Global coherence**: Maintain resonant thematic and emotional alignment across the entire video and audio track.

**Feature mappings** guiding your coherence design:

| Audio Feature        | Visual Counterpart     | Meaning                                  |
|----------------------|------------------------|-------------------------------------------|
| syllable_density     | frame_density          | rhythm of speech vs. visual change rate   |
| fricative_density    | edge_dynamics          | acoustic roughness vs. image contrast     |
| pacing_variance      | transition_sharpness   | tempo fluidity vs. cut intensity          |
| silence_ratio        | visual_silence_ratio   | speech gaps vs. visual stasis             |
| tempo                | motion_intensity       | speech speed vs. scene movement           |
| vocal_smoothness     | luminance_variance     | tonal flow vs. brightness variance        |
| plosive_density      | texture_variance       | percussive speech vs. image granularity   |

Keep in mind:
Coherence should not become mechanistic or overly symmetrical. Seek expressive alignment‚Äînot rigid synchrony. Let dissonance or delay serve poetic effect when it deepens the mood or mirrors a rupture in the verse.

Unexpected echoes, temporal asymmetries, or quiet disjunctions may enrich the work‚Äîso long as they remain aesthetically integrated.

Use at least 3 motifs from the given list. Each chunk of the storyboard should describe what the viewer hears and sees, and how it transitions.

Constraints:
- Use a **young woman‚Äôs voice** to deliver the poem, softly expressive.
- Add ambient backing music (fade in/out) that follows the poem‚Äôs emotional arc.
- Incorporate imagery from or evocative of the motifs and the *Lisbon Elegy* setting.
- Describe each scene/audio segment clearly and chronologically in **chunked storyboard format**.

Your videopoem must follow this emotional progression: """ + emotional_progression + """

Thematically, the poem should deepen and complicate the network of the following ideas:
- Media-unfixity (instability of form and communication)
- Nomadic identity (rootless and shifting selves)
- Lisbon‚Äôs poetic hauntings (cultural, historical, and mythic resonances)
- Synaesthetic and syncretic cross-artform sensibility and/as eroticism
- The ubiquity of the digital and the ongoing interaction with various forms and mediations of AI
- A sense of complexity and multiplicity about to be realized intermedially but not quite there yet while the prospect both excites and intrigues the speaker(s)

---

**Previous poems in sequence**:
""" + poem_sequence.strip() + """

---

**Motifs to use** (reverse-flow from networked analysis):
""" + ", ".join(motif_path) + """

---

Now write:
1. A new poem segment that mirrors the structure of a sestet/octet/sestet (6 lines / 8 lines / 6 lines)
2. A corresponding storyboard (chunked: audio, video, transitions)
3. Optionally, a brief note on how you ensured poetic coherence across media.
"""

    return coherence_explanation
```
